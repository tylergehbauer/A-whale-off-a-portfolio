# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
csv_whale = Path('Resources/whale_returns.csv')

#pd.read_csv reads csv file using pandas
whale_returns_df = pd.read_csv(csv_whale, index_col="Date", infer_datetime_format=True, parse_dates=True)
#while reading the whale_returns.csv, I indexed the 'Date' column in same function via index_col=
#infer_datetime_format puts the dates in the 2015-03-02 format, which we'll use so all data sets have the same row index 

whale_returns_df = whale_returns_df.sort_index()
#this sorts the date from past to present, good for comprehension

whale_returns_df.head() #to check our data frame
```

```python
# Count nulls
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
whale_returns_df.dropna(inplace = True) #inplace = True makes it so the nulls are dropped
#everytime we call the whale_returns_df dataframe
```

```python
#Going to check that nulls were dropped
whale_returns_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
#literally the same exact steps as above but with a different csv file and data frame
csv_algo = Path('Resources/algo_returns.csv')

algo_returns_df = pd.read_csv(csv_algo, index_col="Date", infer_datetime_format=True, parse_dates=True)

algo_returns_df = algo_returns_df.sort_index() #past to present

algo_returns_df.head() 
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df.dropna(inplace = True) #same as above
```

```python
#check if nulls were dropped
algo_returns_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
#again, same exact method as previous 2 data frames when reading in the csv file and creating data frame
csv_sp500 = Path('Resources/sp500_history.csv')

sp500_history_df = pd.read_csv(csv_sp500, index_col="Date", infer_datetime_format=True, parse_dates=True)

sp500_history_df = sp500_history_df.sort_index() #past to present

sp500_history_df.head() 
```

```python
# Check Data Types
sp500_history_df.dtypes #.dtypes is a function that checks for data types
```

```python
# Fix Data Types, we want close to be a float before we calculate the Daily Returns
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$', '')
#this function looks at the 'Close' column and replaces every instance of a '$' with nothing.

sp500_history_df['Close'] = sp500_history_df['Close'].astype('float')
#this function takes the 'Close' column and changes the data type of every entry in this column to a float. 

sp500_history_df.head() 
```

```python
#check the close column consist of float's
sp500_history_df.dtypes
#and we are good to do some calculations!
```

```python
# Calculate Daily Returns
sp500_daily_returns = sp500_history_df.pct_change()
#.pct_change() calculates Daily Returns
sp500_daily_returns
```

```python
# Drop nulls
sp500_daily_returns.dropna(inplace = True)
```

```python
#check if nulls are gone
sp500_daily_returns.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_daily_returns = sp500_daily_returns.rename(columns = {'Close':'S&P 500'})
#.rename renames the column to 'S&P 500' using the dictionary method 
sp500_daily_returns
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.

#we'll use the concat function and concat by columns since these dataframe's share same row index's
combined_df = pd.concat([whale_returns_df, algo_returns_df, sp500_daily_returns], axis="columns", join="inner")

combined_df
#combined_df is all portfolio's daily returns data frame
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
#annualized means multiplying by sqaure root of 252
annual_std = combined_df.std() * np.sqrt(252)
annual_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_std_21 = combined_df.rolling(21).std() #.rolling() is function that lets you calc rolling std/avg
#put the window/ amount of days inside the rolling function parenthesis

# Plot the rolling standard deviation
rolling_std_21.plot(figsize = (20,10), title = '21-Day Rolling STD')
```

### Calculate and plot the correlation

```python
# Calculate the correlation
combined_corr = combined_df.corr()

# Display de correlation matrix
combined_corr
```

```python
#optional, but I want to see a heat map of these correlations 
sns.heatmap(combined_corr, vmax = 1.0, vmin = -1.0)
#darker means less correlated 
#algo 1 seems to be correlating the least between the other portfolios 
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
##I'm going to choose the 'PAULSON & CO.INC.' portfolio for the Beta

# Calculate covariance of a single portfolio (with 60 rolling window to calc 60 day moving beta, via directions) 
covariance = combined_df['TIGER GLOBAL MANAGEMENT LLC'].rolling(60).cov(combined_df['S&P 500'])
#.cov calcs the covariance between 2 varibales/ columns

# Calculate variance of S&P 500 (with 60 day rolling window because of directions)
sp500_variance = combined_df['S&P 500'].rolling(60).var()

# Computing beta
beta = covariance / sp500_variance

# Plot beta trend
beta.plot(figsize = (20,10), title = '60 Day Rolling Beta for Tiger Global Management LLC')
#beta measures volatility of an individal stock compared to the volatility of the market(s&p500)
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
combined_df.ewm(halflife = 21).mean().plot(figsize = (30,15)) #I used .mean() since asked for the average
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratio = (combined_df.mean()*252)/ (annual_std) #I used annual_std variable to save time
#recall: annual_std = combined_df.std()*np.sqrt(252) which is the annualized STD
```

```python
sharpe_ratio
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio.plot(kind = 'bar', title = "Sharpe Ratios")

#Sharpe Ratios calculates the Return-to-Risk ratio. Higher sharpe ratio means more reward for risk
#higher the Sharpe Ratio means more returns
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock
#my first stock is Netflix (aquired google sheets)
#Do the same steps we did earlier. Reading in the csv file with pandas and setting the 'Date' columm as the index in same command
#Also making the index into the datatime format as well

csv_nflx = Path("Resources/NFLX_stockdata.csv")
nflx_df = pd.read_csv(csv_nflx, index_col='Date', infer_datetime_format=True, parse_dates=True)
nflx_df = nflx_df.sort_index()
nflx_df
#Made it such that each of the stocks I gathered through Google Sheets have the same dates.
```

```python
# Reading data from 2nd stock
#my second stock is Disney(aquired google sheets)
csv_disney = Path("Resources/Disney_stockdata.csv")
disney_df = pd.read_csv(csv_disney, index_col = 'Date', infer_datetime_format=True, parse_dates=True)
disney_df = disney_df.sort_index()
disney_df
```

```python
# Reading data from 3rd stock
#my third stock is AMC(aquired google sheets)
csv_amc = Path("Resources/AMC_stockdata.csv")
amc_df = pd.read_csv(csv_amc, index_col = 'Date', infer_datetime_format=True, parse_dates=True)
amc_df = amc_df.sort_index()
amc_df
```

```python
# Combine all stocks in a single DataFrame
#Combine data via rows, this is so we can organize the data with the .pivot_table command
all_stocks = pd.concat([nflx_df, disney_df, amc_df], axis = 'rows', join='inner')
all_stocks
```

```python
# Reset Date index
all_stocks = all_stocks.reset_index()
all_stocks.head()
```

```python
# Reorganize portfolio data by having a column per symbol
my_portfolio = all_stocks.pivot_table(values ='Close', index ='Date', columns='Symbol')
#.pivot_table let's us REORGANIZE (via rows and columns) the data we concated 
#values = 'Close' are the values that will be displayed in each column for each of the 'Symbols'(DIS, NFLX, AMC)
#index makes the index the dates agaiin
my_portfolio
```

```python
# Calculate daily returns
my_daily_returns = my_portfolio.pct_change() #use .pct_change() to calc daily returns 

# Drop NAs
my_daily_returns = my_daily_returns.dropna()
```

```python
#check is all null values are dropped
my_daily_returns.isnull().sum()
```

```python
# Display sample data
my_daily_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3] #each weight is 1/3 such that AMC, DIS, and NFLX hold the same amount of weight

# Calculate portfolio return
my_portfolio_returns = my_daily_returns.dot(weights)
#the .dot function is used to multiply each weight by the corresponding stock daily return
#basically combines them to see how all 3 stocks were perform if they were all weighted equally in my portfolio of stocks

# Display sample data
my_portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
#recall that combined_df is the dataframe with the orignal returns 
combined_df["My Portfolio Returns"] = my_portfolio_returns 

#I have a null value for the first and last row, want to check to make sure
combined_df.isnull().sum()
```

```python
#this will drop the 2 rows with null values, and inplace = True means it stays true one we run this cell. 
combined_df.dropna(inplace = True)
```

```python
#want to just confirm that there is no more null values\
combined_df.isnull().sum()
```

```python
#want to display our results to see if my portfolio is included
combined_df
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)

#I'm going to use the .loc to shorten the data for my analysis
#When graphing the data over 5 years, the graph was hard to understand

#So this data frame will contain the daily returns for each stock from 2017 to 2019-04-22
new_combined_df = combined_df.loc['2017-01-01':'2019-04-22']
new_combined_df
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
my_annualized_std = new_combined_df.std() * np.sqrt(252)
my_annualized_std
#recall we dropped multiple rows, so the annualized STD from before for the orignal stocks are different

#also, my portfolio seems very risky based on this STD
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
my_rolling_std = new_combined_df.rolling(21).std()
# Plot rolling standard deviation
my_rolling_std.plot(figsize = (30,15), title = '21 Day Rolling STD with My Portfolio')
#You can visually compare the rolling standard deviations for each portfolio 
#my portfolio tends to have a high standard deviation
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
my_corr = new_combined_df.corr()
my_corr
```

```python
#to visually see how correlated each portolfio is with one another 
sns.heatmap(my_corr, vmax = 1.0, vmin = -1.0)
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate and plot Beta
my_covariance = new_combined_df["My Portfolio Returns"].rolling(60).cov(new_combined_df["S&P 500"])
my_variance = new_combined_df["S&P 500"].rolling(60).var()
my_beta = (my_covariance/ my_variance)
#used same steps as before, except added .rolling() for rolling 60-day beta 
#Using the .cov to compare my portfolio returns to the S&P 500 

#plot my_beta
my_beta.plot(figsize = (20,10), title = "My Portfolio Beta")
#based on this graph, my portfolio tends to have a high beta, which means more risk with a chance of high returns. 
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualized Sharpe Ratios

#recall earlier we calculated my_annualized_std = new_combined_df.std() * np.sqrt(252)
#that can be place in the denominator to save time
new_sharpe_ratios = (new_combined_df.mean() * 252) / (my_annualized_std)

new_sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
new_sharpe_ratios.plot(kind='bar', title = 'Sharpe Ratios')
```
